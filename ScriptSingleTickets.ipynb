## Python Script to Extract Values for Single Tickets
this code loops over different event names to show the most common price per price section in the concert hall. 
It takes Single Tickets, excluding add-ons and subscribers. 

```python
import numpy as np
import pandas as pd

!pip install numpy pandas matplotlib

import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('/Users/sofialozano/Library/CloudStorage/OneDrive-EmoryUniversity/EMORY FALL 2024/ECON 540 WRITING COMMUNICATION/Project 2 ASO/ASO_data/FY24 Regular Tickets.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
from tabulate import tabulate
```

```python
df_filtered = df[df['price_code'] != 'AAO']
df_filtered.head()
```

```python
df = df_filtered[df_filtered['price_code_type'] == 'Single Ticket']
df.head()
```

```python
# Filter the dataframe for the specific event_name
df_event = df[df['event_name'] == '241019SE']

# Group by price_section and find the most common purchase_price
most_common_price = df_event.groupby('price_section')['purchase_price'].agg(lambda x: x.value_counts().idxmax()).reset_index()

# Display the result

print("Most Common Purchase Price by Price Section for Event: 241019SE")
print(tabulate(most_common_price, headers='keys', tablefmt='fancy_grid'))
```

```python
# Filter the dataframe for the specific event_name
df_event = df[df['event_name'] == '241019SE']

# Calculate the weighted average purchase_price for each price_section
weighted_avg_price = df_event.groupby('price_section').apply(
    lambda x: np.average(x['purchase_price'], weights=x['num_seats'])
).reset_index()

# Rename the columns for clarity
weighted_avg_price.columns = ['price_section', 'weighted_avg_purchase_price']

# Display the result
print("Weighted Average of Prices for Event: 241019SE")
print(tabulate(weighted_avg_price, headers='keys', tablefmt='fancy_grid'))
```

```python
# List of event names
event_names = [
    "241019SE", "241021SE", "241102SE", "241104SE", "241109SE", "241111SE", "241116SE", 
    "241118SE", "241119SM", "241130SE", "241202SE", "240118SE", "240120SE", "240125SE", 
    "240127SE", "240201SE", "240203SE", "240215SE", "240216SE", "240229SE", "240302SE", 
    "240314SE", "240316SE", "240404SE", "240405SE", "240411SE", "240413SE", "240418SE", 
    "240420SE", "240425SE", "240426SE", "240502SE", "240504SE", "240516SE", "240518SE", 
    "240530SE", "240601SE", "240606SE", "240608SE", "240613SE", "240615SE", "240616SM"
]

# Initialize an empty dataframe to store the results
most_common_prices_table = pd.DataFrame()

# Loop through each event name and calculate the most common purchase price by price section
for event in event_names:
    df_event = df[df['event_name'] == event]
    most_common_price = df_event.groupby('price_section')['purchase_price'].agg(lambda x: x.value_counts().idxmax()).reset_index()
    most_common_price.columns = ['price_section', event]
    
    if most_common_prices_table.empty:
        most_common_prices_table = most_common_price
    else:
        most_common_prices_table = pd.merge(most_common_prices_table, most_common_price, on='price_section', how='outer')

most_common_prices_table.head()
# Display the result
#print("Most Common Purchase Price by Price Section for Classical Regular Events")
#print(tabulate(most_common_prices_table, headers='keys', tablefmt='fancy_grid'))
```

```python
most_common_prices_table.to_csv('/Users/sofialozano/Downloads/classical_most_common_prices.csv', index=False)
```

```python
# Initialize an empty dataframe to store the results
weighted_avg_prices_table = pd.DataFrame()

# Loop through each event name and calculate the weighted average purchase price by price section
for event in event_names:
    df_event = df[df['event_name'] == event]
    weighted_avg_price = df_event.groupby('price_section').apply(
        lambda x: np.average(x['purchase_price'], weights=x['num_seats'])
    ).reset_index()
    weighted_avg_price.columns = ['price_section', event]
    
    if weighted_avg_prices_table.empty:
        weighted_avg_prices_table = weighted_avg_price
    else:
        weighted_avg_prices_table = pd.merge(weighted_avg_prices_table, weighted_avg_price, on='price_section', how='outer')

most_common_prices_table.head()
# Display the result
#print("Weighted Average Purchase Price by Price Section for Each Event")
#print(tabulate(weighted_avg_prices_table, headers='keys', tablefmt='fancy_grid'))
```
